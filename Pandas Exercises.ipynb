Exercise 5: Aggregating Data and GroupBy Load the housing data into a DataFrame. Group the data by the 'grade' column (house grade) and calculate advanced statistics for each grade:

Mean price
Median square footage of living space
Maximum number of bedrooms
Minimum number of bathrooms
Display the results as a DataFrame sorted by grade

```python
grade_stats = df_h.groupby('grade').agg({
    'price': 'mean',
    'sqft_living': 'median',
    'bedrooms': 'max',
    'bathrooms': 'min'
}).reset_index()

grade_stats_sorted = grade_stats.sort_values(by='grade')

print("Advanced Statistics for Each Grade:")
print(grade_stats_sorted)
```

Exercise 6: Custom Aggregations Load the housing data and create a custom aggregation function that calculates the average price per square foot of living space for each unique zipcode. Apply this custom aggregation to the data and display the results in a sorted DataFrame.

```python
import pandas as pd

def avg_price_per_sqft(df):
    return (df['price'] / df['sqft_living']).mean()

zipcode_avg_price_per_sqft = df_h.groupby('zipcode').apply(avg_price_per_sqft).reset_index(name='avg_price_per_sqft')

zipcode_avg_price_per_sqft_sorted = zipcode_avg_price_per_sqft.sort_values(by='avg_price_per_sqft')

print("Average Price per Square Foot of Living Space for Each Unique Zipcode:")
print(zipcode_avg_price_per_sqft_sorted)
```
